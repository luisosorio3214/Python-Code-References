```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

## Distribution Plot

```python
# Univariate distribution
sns.histplot(crash_df['not_distracted'], kde = True, bins = 25)
```

## Joint Plot 

```python
# compare 2 distributions

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'kde')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex')
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plots

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.pairplot(tips_df, hue = 'sex', palette = 'Blues')
```

## Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

## Styling 

```python
sns.set_style('white')  # white, whitegrid, dark, ticks, darkgrid

plt.figure(figsize = (8,4))

# change font
sns.set_context('paper', font_scale = 1.4) # talk, poster

sns.jointplot(x = 'speeding',y = 'alcohol', data = crash_df, kind = 'reg')

# default is set to false, border lines on graph
sns.despine(left = False, bottom = False)
```

# Categorical plot

## Bar Plot

```python
sns.barplot(x = 'sex', y='total_bill', data = tips_df, estimator = np.std)
# the bar represents the variance
# by default is aggregated by the mean, can change to np.std, np.cov, np.median, or your own formula
```

## Count Plot

```python
sns.countplot(x = 'sex', data = tips_df)
```

## Box Plot

```python
# compare different variables
# shows quartiles of data

sns.boxplot(x = 'day',y='total_bill', data = tips_df, hue ='sex')
# line in the box represents the median and extends 1 std above 

# move the legend 
plt.legend(loc=0)
```

 ## Violin Plot

```python
# combination of boxplot and kde plot
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', split = True)

# uses kde estimation to create the plot
```

## Strip Plot

```python
plt.figure(figsize = (8,5))

# creates a scatter plot, where one variable is categorical 
sns.stripplot(x= 'day',y = 'total_bill',data = tips_df, jitter = True, hue = 'sex', dodge = True)
# dodge seperates the points from hue
# jitter seperates the data a bit more 
```

## Swarm Plot

```python
sns.violinplot(x='day',y='total_bill', data = tips_df)
sns.swarmplot(x='day',y='total_bill', data = tips_df,color = 'white')
```

```python
sns.swarmplot(x='day',y='total_bill', data = tips_df)
```

# Palettes 

```python
plt.figure(figsize = (8,6))

sns.set_style('dark')
sns.set_context('talk')

sns.stripplot(x = 'day',y='total_bill',data = tips_df, hue='sex', palette = 'afmhot')

# upper right: 1, upper left: 2, lower left: 3, lower right: 4, 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc = 1)
```

## Matrix Plot

### Heat Map

```python
plt.figure(figsize = (8,6))

sns.set_context('paper', font_scale = 1.4)

crash_mx = crash_df.corr(numeric_only = True)

sns.heatmap(crash_mx, annot = True, cmap = 'Blues')
```

```python
flights= sns.load_dataset('flights')
flights = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
flights
```

```python
sns.heatmap(flights, cmap = 'Blues', linecolor = 'white',linewidth =1)
```

## Pair Grid

```python
iris = sns.load_dataset('iris')
iris_g =  sns.PairGrid(iris, hue = 'species')
#iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g =  sns.PairGrid(iris, hue = 'species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g =  sns.PairGrid(iris, hue = 'species', 
                      x_vars = ['sepal_length','sepal_width'],
                      y_vars = ['petal_length','petal_width'])

iris_g.map(plt.scatter)

iris_g.add_legend()
```

 ## Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df, col ='time', row = 'smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col ='time', row = 'smoker')
tips_fg.map(plt.hist, 'total_bill', bins = 8)
```

```python
tips_fg = sns.FacetGrid(tips_df, col ='time', hue = 'smoker', height = 4, aspect = 1.3,
                       col_order = ['Dinner','Lunch'], palette = 'Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor = 'w')
```

```python
kws = dict(s=50,linewidth = .5, edgecolor = 'w')
tips_fg = sns.FacetGrid(tips_df, col = 'sex', hue = 'smoker',height =4, aspect = 1.3,
                        hue_order = ['Yes','No'],
                        hue_kws = dict(marker = ['^','v']))
tips_fg.map(plt.scatter, 'total_bill','tip',**kws)
```
