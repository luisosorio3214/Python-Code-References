### Load the necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Basic Plot

```python
x = [0,1,2,3,4] # 1-d array 
y = [0,2,4,6,8] #1-d array

#Resize your Graph
plt.figure(figsize=(5,3),dpi = 100) 
# dpi pixels per inch, (5,3) then 1500 per 900 pixels

## Line 1

#Keyword argument Notation
#plt.plot(x,y, label='2x', color = 'red', linewidth = 2, marker = '.',linestyle = '--', markersize =10,markeredgecolor='green')

# Can use short hand notation for plt.plot
# fmt = '[color][marker][line]'
plt.plot(x,y,'b^--', label='2x')


## Line two 

#Select interval of what we want to plot
x2 = np.arange(0,4.5,0.5)
#print(x2)

#Plot part of the graph as a line
plt.plot(x2[:6],x2[:6]**2,'r',label = 'x^2')

#Plot reminder of the graph as dashed lines
plt.plot(x2[4:],x2[4:]**2,'r--')

#Adding a title, specify font parameter 
plt.title("Our First Graph!", fontdict = {'fontname': "comic sans MS", 'fontsize': 20})

#Adding label for X & Y
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X & Y tick marks (scaling of your graph)
# Automatically changes the sizing of the graph
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Plots the legend but first needs a label in plt.plot
plt.legend()

#how to save graph, same place where this notebook is saved
# saving dpi = 300 is good so graph has a high resolution
#plt.savefig('mygraph.png',dpi=300)

# To get rid of the line above if you only run the above code. Show Plot 
plt.show()
```

### HIstograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef') # for color go to colorpicker

plt.xticks(bins)
plt.yticks()

plt.title('Distribution of Player Skills in FIFA 2018')
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0] #store the len(left)

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0] #store the len(right)


#len(left), len(right)

labels = ['Left','Right']
colors = ['#abcedf','#aabbcc']

plt.pie([left,right],labels = labels,colors=colors, autopct = '%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart #2

```python
fifa.Weight
# Notice its in lbs! want to convert into integer int
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

```python
fifa.Weight[0]
# We have turned it into integers
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]

plt.figure(figsize=(15,8))
plt.style.use('ggplot')
plt.title('Weight Distribution of FIFA Players (in lbs)')
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','Over 200']
explode = (.4,.2,0,0,.4)

plt.pie(weights,labels=labels, autopct ='%.2f %%', pctdistance = 0.8, explode=explode )

plt.show()
```

### Box and Whisker Plot

```python
plt.style.use('default')

plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']


labels = ['FC Barcelona','Real Madrid', 'NE Revolution'] # NE New England

boxes = plt.boxplot([barcelona,madrid,revs], labels=labels,patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    #set edge color
    box.set(color='#4286f4', linewidth = 2)
    
    #change fill color 
    box.set(facecolor='#e0e0e0')

    
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
